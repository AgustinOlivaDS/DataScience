### Serie de tiempo
**Indice igual al año de los juegos olímpicos**

```python
import pandas as pd
import numpy as np

# leo el archivo con encoding latin_1 porque tiene caractéres fuera el utf-8
archivo = pd.read_csv("files/Summer-Olympic-medals-1976-to-2008.csv", encoding="latin_1") 

# explorar DataFrame
# archivo.head(6)
# print(archivo.dtypes)
# print(archivo.shape)
# print(archivo.columns)

# completo los NAN
archivo["Year"] = archivo["Year"].fillna(0)
years = archivo["Year"].astype(np.int32)
archivo.index = years
archivo.head()
```

### Separo los nombre de los apellidos por coma

```python
archivo[["LastName", "Name", "Name2"]] = archivo["Athlete"].str.split(",", expand=True)

archivo = archivo[['City', 'Year', 'Sport', 'Discipline', 'Event', 'Athlete', 'Gender',
       'Country_Code', 'Country', 'Event_gender', 'Medal', 'LastName', 'Name']]
archivo.head()
```

### Resumen de medidas, qué país ganó mas medallas

```python
# Top muestra el país con más medallas e freq la cantidad
print(archivo['Country'].describe())
```

### Tabla con la cantidad de medallas de los hombres por año

```python
men_medals = archivo[(archivo["Gender"] == "Men")]
men_medals["Year"].astype(np.int32).value_counts()
```
